### Soal 1: Bitcoin Price

Menggunakan Line Chart karena dapat menampilkan trend naik maupun turun dari data harga bitcoin pada periode tertentu.
Tahun 2018 menggunakan line chart berwarna biru menunjukan adanya tren penurunan, sedangkan tahun 2019 menggunakan line chart berwarna orange menunjukan adanya tren kenaikan. Jadi dapat disimpulkan bahwa tahun 2019 memberikan pengembalian/keuntungan yang lebih baik daripada tahun 2018.

### Soal 2: Permen

Menggunakan Pie Chart karena dapat menampilkan persentase dari data yang digunakan.
Persentase peluang memilih permen Kopiko sebesar 15%.

```python
import matplotlib.pyplot as plt

nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
jumlah_permen = [52, 39, 78, 13, 78]
warna = ('#00BFFF', '#CD853F', '#FFE4C4', '#EE82EE', '#00FF7F')
highlight = (0,0.2,0,0,0)

plt.title('Peluang memilih permen Kopiko\n')
plt.pie(jumlah_permen, labels = nama_permen, autopct = '%1.1f%%', colors = warna, explode = highlight, shadow = True)
plt.show()
```

### Soal 3: Makanan

Menggunakan Bar Chart karena dapat menampilkan data dengan mengurutkan dari penjualan tertinggi ke penjualan terendah ataupun sebaliknya.
3 makanan penutup yang harus disingkirkan yaitu Puding Vanila, Pastel, dan Kue Wajik.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datapenjualan_makananpenutup = {
    'Donat': 14,  
    'Pastel': 5,  
    'Kue Coklat': 12,  
    'Ice Cream': 19,  
    'Puding Vanila': 6,  
    'Brownies': 8,  
    'Puding Strawberi': 12,  
    'Puding Coklat': 9,  
    'Ice Cream Nutela': 10, 
    'Kue Coklat-Keju': 17,  
    'Kue Wajik': 2,  
    'Kue Sus': 9,  
    'Mochi': 13, 
}

makanan = list(datapenjualan_makananpenutup.keys())
data_penjualan = list(datapenjualan_makananpenutup.values())

df = pd.DataFrame({'Makanan':makanan, 'Data':data_penjualan})
df.sort_values(by='Data', inplace = True, ascending = False)

kor_x = np.arange(len(df))

warna = ['#0000FF' for _ in range(len(df))]
warna[-3] = '#FF0000'
warna[-2] = '#FF0000'
warna[-1] = '#FF0000'

plt.figure(figsize=(20,10))
plt.bar(kor_x, df['Data'], tick_label = df['Makanan'], color = warna)
plt.xticks(rotation=90)
plt.ylabel('Data Penjualan')
plt.title('Data Penjualan Menu Makanan Penutup')
plt.show()
```
